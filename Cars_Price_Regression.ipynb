```
cars_df.drop(axis=1, columns=['car_ID'], inplace=True)
cars_df.drop(axis=1, columns=['CarName'], inplace=True)
cars_df
```

```
cars_df.columns = ['symboling','is_gas', 'is_turbo', 'is_four_door', 'carbody', 'drivewheel',
       'is_front_engine', 'wheelbase', 'carlength', 'carwidth', 'carheight',
       'curbweight', 'enginetype', 'cylindernumber', 'enginesize',
       'fuelsystem', 'boreratio', 'stroke', 'compressionratio', 'horsepower',
       'peakrpm', 'citympg', 'highwaympg', 'price']
cars_df.is_gas.replace({'gas':1, 'diesel':0}, inplace=True)
cars_df.is_turbo.replace({'turbo':1, 'std':0}, inplace=True)
cars_df.is_four_door.replace({'four':1, 'two':0}, inplace=True)
cars_df.is_front_engine.replace({'front':1, 'back':0}, inplace=True)
cars_df
```

```
cars_df.cylindernumber.replace({'four':4, 'six':6, 'three':3, 'twelve':12, 'two':2, 'eight':8, 'five':5},inplace=True)
```

```
cars_df.drop('carbody', axis=1, inplace = True)
cars_df.drop('drivewheel', axis=1, inplace = True)
cars_df.drop('enginetype', axis=1, inplace = True)
cars_df.drop('fuelsystem', axis=1, inplace = True)
```

```
pd.set_option("display.precision", 3)
corr_df = cars_df.corr()
corr_df
```

```
fig = go.Figure()
fig.add_traces(go.Heatmap(
    z=corr_df, 
    x=corr_df.columns, 
    y=corr_df.columns,
    zmax=1, 
    zmin=-1
))
fig.update_layout({
    'title':"cars_df corr heatmap"
})
```

```
for i in range(len(corr_df.columns)-1):
  print(corr_df.columns[i], 'is correlated to: ')
  for j in range(len(corr_df.columns)-1):
    if abs(corr_df.iloc[i,j])>=0.85 and i is not j:
      print(corr_df.columns[j], end=', ')
  print('\n----------------------')
```

```
cars_df.drop('highwaympg', axis=1, inplace = True)
cars_df.drop('compressionratio', axis=1, inplace = True)
cars_df.drop('wheelbase', axis=1, inplace = True)
cars_df.drop('curbweight', axis=1, inplace = True)
```

```
corr_df = cars_df.corr()['price'] # creates new correlation df, and keeps only the 'priceK' column
corr_df = corr_df.iloc[0:-1] # drop the last line, because we dont need the correlation between 'priceK' and 'priceK' (that will be 1)
correlated = corr_df.loc[abs(corr_df)>0.66] # abs() returns absolute value. and using loc[] we can slice rows with values that are greater then 0.66
correlated
```

```
from sklearn.model_selection import train_test_split
X=cars_df[['enginesize']]
y=cars_df.price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
train_df=pd.merge(left=X_train, right=y_train, left_index=True, right_index=True)
test_df=pd.merge(left=X_train, right=y_train, left_index=True, right_index=True)
train_df.head(10)
```

```
fig = go.Figure()
fig.add_trace(go.Box(name='enginesize outliers', y=train_df.enginesize))
fig.show()
```

```
from sklearn.preprocessing import StandardScaler
x_train_scaler = StandardScaler()
x_test_scaler = StandardScaler()
y_train_scaler = StandardScaler()
y_test_scaler = StandardScaler()

X_train_scaled = x_train_scaler.fit_transform(train_df[['enginesize']])
X_test_scaled = x_test_scaler.fit_transform(test_df[['enginesize']])
y_train_scaled = y_train_scaler.fit_transform(train_df[['price']])
y_test_scaled = y_test_scaler.fit_transform(test_df[['price']])

```

```
from sklearn.linear_model import LinearRegression
#define model
lm = LinearRegression()
lm.fit(X_train_scaled,y_train_scaled)
slope = lm.coef_[0]
intercept = lm.intercept_
print('Coefficients: \n', "slope =", slope , ",  Intercept=",intercept,"\n" )
print("The regression equation is:" ,"Price = [" ,intercept,"] + [",slope,"] * curbweight")
```

```
#use lm.predict(X_train) for all the X data from the train set
price_by_train=lm.predict(X_train_scaled)
fitted_price_train= y_train_scaler.inverse_transform(price_by_train)
predicted_train_price =round(pd.Series(fitted_price_train[:,0],index=X_train.index,name='Predicted_train_price'),ndigits=2)
train_df=pd.merge(left=train_df,right=predicted_train_price,left_index=True,right_index=True)
train_df.head(10)
```

```
price_by_test=lm.predict(X_test_scaled)
fitted_price_test= y_test_scaler.inverse_transform(price_by_test)
predicted_test_price = round(pd.Series(fitted_price_test[:,0],index=X_train.index,name='Predicted_test_price'),ndigits=2)
test_df=pd.merge(left=test_df,right=predicted_test_price,left_index=True ,right_index=True)
test_df
```

```
#1 Cost functions
from sklearn import metrics
print("MSE:",metrics.mean_squared_error(train_df.price, train_df.Predicted_train_price))
print("RMSE:",np.sqrt(metrics.mean_squared_error(train_df.price, train_df.Predicted_train_price)))
print("MAE:",metrics.mean_absolute_error(train_df.price, train_df.Predicted_train_price))




```

```
from sklearn.metrics import r2_score
r2_score(train_df.price, train_df.Predicted_train_price)
```

```
r2_score(test_df.price, test_df.Predicted_test_price)
```
