```
df= wine_df[wine_df['pH'] != 0]
df.shape[0]
df.pH.unique()
```

```
missing = df.isnull()
missing.sum()
```

```
df=df.dropna()
df_sum=df.isnull()
df_sum.sum()
```

```
df.sort_values('alcohol',ascending=True, inplace=True,ignore_index=True)
```

```

fig = go.Figure()
fig.add_traces(go.Heatmap(
    z=df.corr(), 
    x= df.columns,  
    y= df.columns,
))
fig.update_layout({
    'title':"Wine DF Dataframe Correlation Heatmap"
})


fig.show()
```

```
df.quality.replace({'good':1,'bad':0}, inplace=True)
```

```
df.quality
```

```
show_quality = df.quality.value_counts()
print(show_quality)
precentage = round(df.quality.value_counts(normalize=True) * 100, ndigits = 1 ) 
print(precentage)
```

```
fig = go.Figure()

fig.add_trace(go.Bar(
    x=['Good', 'Bad'],
    y=show_quality,
    marker_color=['red','green'],
    text=precentage ,
    textposition='outside',
)) 

fig.update_layout(
    title='Quality', 
    yaxis_title='N',
    )
fig.show()
```

```
X = df.drop('quality',axis=1)
y = df['quality']
```

```
#X
#y
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1234)
```

```
from sklearn.tree import DecisionTreeClassifier
```

```
dtree = DecisionTreeClassifier(criterion="entropy",max_depth=9)
```

```
dtree.fit(X_train,y_train)
```

```
probability = dtree.predict_proba(X_test)
prediction = dtree.predict(X_test)
#probability
y_test_prob_0 = pd.Series(probability[:,0], name ='probability_0', index=X_test.index)
y_test_prob_1 = pd.Series(probability[:,1], name ='probability_1', index=X_test.index)
```

```
dtree.predict_proba(X_test[:10])
```

```
from IPython.display import Image  
from io import StringIO 
from sklearn.tree import export_graphviz
import pydot 
```

```
features = list(X.columns)
dot_data = StringIO()  
export_graphviz(dtree, out_file=dot_data,feature_names=features,filled=True,rounded=True, node_ids=True)
graph = pydot.graph_from_dot_data(dot_data.getvalue())  
Image(graph[0].create_png()) 
```

```
print(dtree.feature_importances_)
```

```
fig = go.Figure()
trace = go.Bar(
    x = dtree.feature_importances_,
    y = features,
    orientation='h'
)
fig.add_trace(trace)
```

```
y_pred = dtree.predict(X_test)
y_pred_df = pd.DataFrame(y_pred ,columns=['y_pred'])
pred_df = pd.DataFrame({"prob_0":y_test_prob_0, "prob_1":y_test_prob_1,"Actual": df.quality,"Prediction":y_pred} , index = y_test.index)
pred_df
```

```
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```
from sklearn.metrics import classification_report
from sklearn import metrics
```

```
fp_rate, tp_rate, thresholds = metrics.roc_curve(y_test,y_test_prob_1)
```

```
fig=go.Figure()

trace=go.Scatter(x=fp_rate, y=tp_rate, mode='lines', hoverinfo='x+y+text', text=thresholds)

fig.add_trace(trace)

fig.update_layout(
    title="ROC curve for wine quality classifier",
    xaxis_title="False Positive Rate (1 - Specificity)",
    yaxis_title="True Positive Rate (Sensitivity/Recall)",
    
)
fig.show()
```

```
AUC = metrics.roc_auc_score(y_test, y_test_prob_1)
print('AUC',AUC)
```

```
from sklearn.preprocessing import binarize
y_pred_08 = binarize(X=[y_test_prob_1], threshold=0.8)[0]
y_pred_08 = pd.Series(y_pred_08)
y_pred_08
```

```
print(classification_report(y_test,y_pred_08))
```

```
from sklearn.model_selection import cross_val_score
auc_scores = cross_val_score(dtree, X, y, cv=30, scoring='roc_auc')
auc_scores
```

```
accuracy_scores = cross_val_score(dtree, X, y, cv=30, scoring='accuracy')
accuracy_scores
```
